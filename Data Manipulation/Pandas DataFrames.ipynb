```python
import pandas as pd
import numpy as np
```

##### Creating a random df

```python
np.random.seed(6) 


boxes = {'X': np.random.randint(10,20,10),
         'Y': np.random.randint(40,100,10),
         'Z': np.random.randint(0,10,10)}
```

```python
df = pd.DataFrame(boxes)
```

```python
df
```

##### Changing column names

```python
labels = ['Variable_A', 'Variable_B', 'Variable_C']
```

```python
df.columns = labels
df
```

```python
df.info()
```
